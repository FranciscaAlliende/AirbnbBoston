```python
# select relevant features
cols =  ['price', 'neighbourhood_cleansed', 'property_type', 'room_type','accommodates', 'bathrooms', 'bedrooms',
         'beds','bed_type','amenities','square_feet', 'cancellation_policy', 'host_since', 'host_response_time',
         'host_response_rate','host_acceptance_rate','host_is_superhost', 'host_identity_verified', 'number_of_reviews',
         'reviews_per_month','review_scores_accuracy','review_scores_cleanliness','review_scores_checkin',
         'review_scores_communication','review_scores_location','review_scores_value']
data = data[cols]
data.head()
```

```python
data.dtypes
```

```python
# remove dollar sign and convert to float
data['price'] = data['price'].replace('[\$,]', '', regex=True).astype(float)
```

```python
# change name column
data = data.rename(columns={'neighbourhood_cleansed': 'neighbourhood'})
```

```python
# host_response_rate, from % to float
data['host_response_rate'] = data['host_response_rate'].replace('[\%,]', '', regex=True).astype(float)

# host_acceptance_rate, from % to float
data['host_acceptance_rate'] = data['host_acceptance_rate'].replace('[\%,]', '', regex=True).astype(float)
```

```python
def DaysPassed(col, df):
    # convert to datetime dtype
    df[col] = pd.to_datetime(df[col])
    
    # date when the dataset was scrapped
    last_scraped = pd.to_datetime("2016-09-07")
    
    # return the difference between the 2 dates in days
    return (last_scraped - df[col]).dt.days 
```

```python
# from date to days since
data['host_since'] = DaysPassed('host_since', data)
```

```python
# Amenities, clean
data['amenities'] = data['amenities'].replace("{", '', regex=True)
data['amenities'] = data['amenities'].replace("}", '', regex=True)
data['amenities'] = data['amenities'].replace("\"", '', regex=True)

# Convert to dummies
A = data['amenities'].str.get_dummies(sep= ",")
A = A.drop('translation missing: en.hosting_amenity_49', 1)
A = A.drop('translation missing: en.hosting_amenity_50', 1)

A.columns
```

```python
# drop amenities columns
data = data.drop('amenities', 1)
```

```python
# concatenate data with amenities
result = pd.concat([data, A], axis=1)
result.head()
```

```python
result.to_csv('data/Boston/result.csv', index=False)
```
